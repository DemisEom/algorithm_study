# [HackerRank] Sherlock and the Valid String

## 구해야 하는 것

- 주어지는 문자열이 규칙에합당한지 아닌지를 판별하는 문제
- 문자열의 모든 문자가 동일한 횟수로 표시되면 유효.
- 또한, 한 문자만 제거하여 위 조건을 만족하면 유효하다고 판단.

## 주어지는 것

- 소문자 알파벳으로 되어 있는 문자열

## 어떻게 풀것인가

- 문자들의 개수를 카운팅. -> 딕셔너리나 리스트를 생성. -> 각케이스의 개수를 확인.
- 케이스는 크게 3가지.
    - 모든 개수가 같거나
    - 한개만 1개 더 많거나
    - 모든 개수가 같은데 하나만 1개이거나

## 내 풀이

```python
def isValid(s):
    keys = sorted(list(set(s)))
    char_dict = { i:0 for i in keys}
    
    for i in range(len(s)):
        char_dict[s[i]] += 1
    cnt_list = list(char_dict.values())
    
#     print(char_dict)
    
    cnt_list.sort(reverse=True)
    
    print(cnt_list)
    cnt_length = len(cnt_list)
    
    if cnt_length == 1:
        return "YES"
    
    if (cnt_list[0]-1 == cnt_list[1]) and (sum(cnt_list[1:])%(len(cnt_list)-1)) == 0:
        return "YES"
    
    if cnt_list[-1] == 1 and sum(cnt_list[:cnt_length-1])==cnt_list[0]*(cnt_length-1):
        return "YES"
    
    if cnt_list[0]*cnt_list == sum(cnt_list):
        return "YES"
    
    return "NO"
```
