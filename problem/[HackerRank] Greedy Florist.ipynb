# [HackerRank] Greedy Florist

## 구해야 하는 것

- 친구들이 부케를 사려고 함.
- 꽃파는 사람은 새로운 고객을 최대한 많이 해서 돈을 벌고자 함.
- 이를 위해서, 원래 가격에 *(1+이전에 구매한 개수) 로 판매를 한다.
- 친구들의 숫자와 꽃의 오리지널 가격들이 주어졌을때 모든 꽃을 구매하기 위한 최소 가격을 구하시오.
    - 3명의 친구, [1,2,3,4]의 오리지널 꽃가격
    - [2,3,4]를 한명씩 구매하고, 한명이 1원짜리를 한번더 구매하면 2원에 살수 있음.
    - 이때 2+3+4+2가되서 11로 최소로 구매가능.

## 주어지는 것

- 구매하는 사람의 숫자
    - 100이하의 상수
- 꽃의 원래 가격
    - 길이는 최대 10^6

## 어떻게 구할것인가

- 비싼것들을 먼저 구매하고 싼것들을 더 구매.
- 이때, 사람의 숫자와 꽃 종류, 길이를 비교해야함.
    - 사람 숫자 >= 꽃 길이
        - 전체를 1번씩 구매
    - 사람 숫자 < 꽃 길이
        - 정렬 시키후 비싼것 먼저 제외.
        - 반복문으로 pop시키면서 값 계산?


## 내 풀이

```python
def getMinimumCost(k, c):
    
    sorted_price = sorted(c)
    
    if k >= len(c): 
        return sum(c)
    else:
        print(sorted_price)
        total_price = sum(sorted_price[len(c)-k:])
        sorted_price = sorted_price[:len(c)-k]
        
        buy = 2
        flag = 0
        print(sorted_price)
        while sorted_price:
            temp_price = sorted_price.pop()
            
            if flag == k:
                flag = 0
                buy += 1
                
            print(flag, buy, temp_price, buy*temp_price)    
            total_price += buy*temp_price
            flag +=1
            
        return total_price   
```
